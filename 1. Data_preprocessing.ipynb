## Change the data format to input into GATNE

```python
df_AI=df_AI[['publn_nr','IPCcodes','publn_date']]
df_FT=df_FT[['publn_nr','IPCcodes','publn_date']]
df_AT=df_AT[['publn_nr','IPCcodes','publn_date']]

import re
def convert_ipc_code(ipc):
  pattern = r'(\b\w+\b)\s+(\w+\b)'
  match = re.search(pattern, ipc)
  if match:
    return match.group(1) + '-' + '0'*(4-len(match.group(2)))+match.group(2)

df_AI['IPCcodes'] = df_AI['IPCcodes'].apply(convert_ipc_code)
df_FT['IPCcodes'] = df_FT['IPCcodes'].apply(convert_ipc_code)
df_AT['IPCcodes'] = df_AT['IPCcodes'].apply(convert_ipc_code)
```

```python
#publn_nr로 그룹핑
df_AI.dropna(inplace=True)
df_AI= df_AI.groupby(['publn_nr', 'publn_date'])['IPCcodes'].apply(list).reset_index()
df_AI['IPCcodes'].apply(lambda x: list(set(x)))

df_FT.dropna(inplace=True)
df_FT= df_FT.groupby(['publn_nr', 'publn_date'])['IPCcodes'].apply(list).reset_index()
df_FT['IPCcodes'].apply(lambda x: list(set(x)))

df_AT.dropna(inplace=True)
df_AT= df_AT.groupby(['publn_nr', 'publn_date'])['IPCcodes'].apply(list).reset_index()
df_AT['IPCcodes'].apply(lambda x: list(set(x)))

df_AT
```

```python
from itertools import combinations

def combination(IPCcodes):
  if len(IPCcodes)>1:
    IPC_combination_list=[list(combination) for combination in combinations(IPCcodes, 2)]
    return IPC_combination_list
  else:
    return None



df_AI['IPCcodes']=df_AI['IPCcodes'].apply(combination)
df_FT['IPCcodes']=df_FT['IPCcodes'].apply(combination)
df_AT['IPCcodes']=df_AT['IPCcodes'].apply(combination)

df_AI = df_AI.explode('IPCcodes')
df_FT = df_FT.explode('IPCcodes')
df_AT = df_AT.explode('IPCcodes')
df_AI.reset_index(drop=True, inplace=True)
df_FT.reset_index(drop=True, inplace=True)
df_AT.reset_index(drop=True, inplace=True)

df_AI.dropna(inplace=True)
df_FT.dropna(inplace=True)
df_AT.dropna(inplace=True)

```

```python
df_AI['IPCcodes'] = df_AI['IPCcodes'].apply(lambda x: sorted(x) if isinstance(x, list) else x)
df_FT['IPCcodes'] = df_FT['IPCcodes'].apply(lambda x: sorted(x) if isinstance(x, list) else x)
df_AT['IPCcodes'] = df_AT['IPCcodes'].apply(lambda x: sorted(x) if isinstance(x, list) else x)

df_AI['IPCcodes'] = df_AI['IPCcodes'].apply(lambda x: "1 " + " ".join(x) + " 1" if isinstance(x, list) else "")
df_FT['IPCcodes'] = df_FT['IPCcodes'].apply(lambda x: "1 " + " ".join(x) + " 1" if isinstance(x, list) else "")
df_AT['IPCcodes'] = df_AT['IPCcodes'].apply(lambda x: "1 " + " ".join(x) + " 1" if isinstance(x, list) else "")
```

```python
def combine_with_count1(series):
    counts = series.value_counts()
    result = []
    for value, count in counts.items():
      result.append(f"1 {value[2:-2]} 1")
    return ", ".join(result)

def combine_with_count2(series):
    counts = series.value_counts()
    result = []
    for value, count in counts.items():
      result.append(f"2 {value[2:-2]} 1")
    return ", ".join(result)

def combine_with_count3(series):
    counts = series.value_counts()
    result = []
    for value, count in counts.items():
      result.append(f"3 {value[2:-2]} 1")
    return ", ".join(result)

df_AI['IPCcodes'] = df_AI.groupby('IPCcodes')['IPCcodes'].transform(combine_with_count1)
df_FT['IPCcodes'] = df_FT.groupby('IPCcodes')['IPCcodes'].transform(combine_with_count2)
df_AT['IPCcodes'] = df_AT.groupby('IPCcodes')['IPCcodes'].transform(combine_with_count3)
df_AI.drop_duplicates(inplace=True)
df_FT.drop_duplicates(inplace=True)
df_AT.drop_duplicates(inplace=True)
```

```python
# 빈 문자열을 가진 행 삭제
df_AI.drop(df_AI[df_AI['IPCcodes'] == ''].index, inplace=True)
df_FT.drop(df_FT[df_FT['IPCcodes'] == ''].index, inplace=True)
df_AT.drop(df_AT[df_AT['IPCcodes'] == ''].index, inplace=True)

#df_AI= df_AI.drop_duplicates()
#df_FT= df_FT.drop_duplicates()
#df_AT= df_AT.drop_duplicates()
```

## Time frame splitting

```python
def year2001_2005(df1):
  df1['year'] = df1['publn_date'].str[:4].astype(int)
  df1_2001_2005 = df1[df1['year'].between(2001, 2005)]
  df1_2001_2005 = df1_2001_2005.drop(columns='year')
  return df1_2001_2005

def year2006_2010(df1):
  df1['year'] = df1['publn_date'].str[:4].astype(int)
  df1_2006_2010 = df1[df1['year'].between(2006, 2010)]
  df1_2006_2010 = df1_2006_2010.drop(columns='year')
  return df1_2006_2010

def year2011_2015(df1):
  df1['year'] = df1['publn_date'].str[:4].astype(int)
  df1_2011_2015 = df1[df1['year'].between(2011, 2015)]
  df1_2011_2015 = df1_2011_2015.drop(columns='year')
  return df1_2011_2015

def year2016_2020(df1):
  df1['year'] = df1['publn_date'].str[:4].astype(int)
  df1_2016_2020 = df1[df1['year'].between(2016, 2020)]
  df1_2016_2020 = df1_2016_2020.drop(columns='year')
  return df1_2016_2020

df2001_2005=pd.concat([year2001_2005(df_AI),year2001_2005(df_FT),year2001_2005(df_AT)])
df2006_2010=pd.concat([year2006_2010(df_AI),year2006_2010(df_FT),year2006_2010(df_AT)])
df2011_2015=pd.concat([year2011_2015(df_AI),year2011_2015(df_FT),year2011_2015(df_AT)])
df2016_2020=pd.concat([year2016_2020(df_AI),year2016_2020(df_FT),year2016_2020(df_AT)])
```

```python
df2001_2005=df2001_2005[['IPCcodes']]
df2006_2010=df2006_2010[['IPCcodes']]
df2011_2015=df2011_2015[['IPCcodes']]
df2016_2020=df2016_2020[['IPCcodes']]
```

```python
def rename_column(df_all):
  df_all.rename(columns={'IPCcodes':'A'}, inplace=True)

rename_column(df2001_2005)
rename_column(df2006_2010)
rename_column(df2011_2015)
rename_column(df2016_2020)
```

```python
df2001_2005 = df2001_2005.drop_duplicates(subset='A', keep='first')
df2006_2010 = df2006_2010.drop_duplicates(subset='A', keep='first')
df2011_2015 = df2011_2015.drop_duplicates(subset='A', keep='first')
df2016_2020 = df2016_2020.drop_duplicates(subset='A', keep='first')
```
